```python
import pandas as pd
df_iris=pd.read_csv(r'C:\Users\paster\Downloads\datasets (3)\datasets\seaborn_datasets\iris.csv')
print(df_iris)
```

```python
# Media della lunghezza dei petali di tutto il dataset
media_petali = df_iris['petal_length'].mean()
print(f"Media della lunghezza dei petali di tutto il dataset: {media_petali:.2f}")
```

```python
# Media della lunghezza dei petali per ogni specie usando groupby()
media_petali_per_specie = df_iris.groupby('species')['petal_length'].mean()
print("Media della lunghezza dei petali per ogni specie:\n", media_petali_per_specie)
```

```python
#  Media, minimo e massimo della larghezza dei sepali per ogni specie usando groupby() e agg()
stats_sepali = df_iris.groupby('species')['sepal_width'].agg(['mean', 'min', 'max'])
print("Statistiche sulla larghezza dei sepali per ogni specie:\n", stats_sepali)
```

```python
import pandas as pd
df_boston=pd.read_csv(r'C:\Users\paster\Downloads\datasets (3)\boston.csv')
df_boston
```

```python
print(df_boston.head())
print(df_boston.columns)
```

```python
#media del tasso di criminalità
media_crim = df_boston['crim'].mean()
print(f"Media del tasso di criminalità: {media_crim:.2f}")
```

```python
#media del numero di stanze
media_rm = df_boston['rm'].mean()
print(f"Numero medio di stanze per abitazione: {media_rm:.2f}")
```

```python
#Analisi della variabile chas
medv_per_chas = df_boston.groupby('chas')['medv'].mean().round(2)
print("Valore medio delle case in zone con CHAS=0 e CHAS=1:\n", medv_per_chas)
```

```python
#correlazione tra lstat e medv
correlation = df_boston['lstat'].corr(df_boston['medv'])
print(f"Correlazione tra LSTAT e MEDV: {correlation:.2f}")
```

```python
#correlazione tra crim e medv\ si deduce che la correlazione è negativa quindi un aumento del tasso di criminalità tende ad associarsi a un valore più basso delle case. 
#Questo può essere spiegato dal fatto che aree più pericolose o degradate spesso hanno prezzi immobiliari più bassi.
correlazione_crim_medv = df_boston['crim'].corr(df_boston['medv'])
print(f"Correlazione tra CRIM e MEDV: {correlazione_crim_medv:.2f}")
```

```python
# Media di RM rispetto a PTRATIO
media_rm_per_ptratio = df_boston.groupby('ptratio')['rm'].mean()
# Media di MEDV rispetto a PTRATIO
media_medv_per_ptratio = df_boston.groupby('medv')['ptratio'].mean()

print("Media di RM rispetto a PTRATIO:\n", media_rm_per_ptratio)
print("\nMedia di MEDV rispetto a PTRATIO:\n", media_medv_per_ptratio)
```

```python
# Correlazioni
corr_rm_ptratio = df_boston['rm'].corr(df_boston['ptratio'])
corr_medv_ptratio = df_boston['medv'].corr(df_boston['ptratio'])

print(f"\nCorrelazione tra RM e PTRATIO: {corr_rm_ptratio:.2f}")
print(f"Correlazione tra MEDV e PTRATIO: {corr_medv_ptratio:.2f}")
#Se RM e PTRATIO sono negativamente correlati, significa che aree con più stanze tendono ad avere un rapporto alunni-insegnanti più basso (migliore).
#Se MEDV e PTRATIO sono negativamente correlati, indica che zone con meno studenti per insegnante tendono ad avere case più costose.
```

```python
# Per esempio, consideriamo due gruppi: RAD <= 5 e RAD > 5
gruppo_bassa_accessibilita = df_boston[df_boston['rad'] <= 5]
gruppo_alta_accessibilita = df_boston[df_boston['rad'] > 5]

# Medie per AGE, RM, TAX nei due gruppi
medie_age_low_rad = gruppo_bassa_accessibilita['age'].mean()
medie_age_high_rad = gruppo_alta_accessibilita['age'].mean()

medie_rm_low_rad = gruppo_bassa_accessibilita['rm'].mean()
medie_rm_high_rad = gruppo_alta_accessibilita['rm'].mean()

medie_tax_low_rad = gruppo_bassa_accessibilita['tax'].mean()
medie_tax_high_rad = gruppo_alta_accessibilita['tax'].mean()

print(f"Età media (age) - rad <=5: {medie_age_low_rad:.2f}, rad >5: {medie_age_high_rad:.2f}")
print(f"Numero medio di stanze (rm) - rad <=5: {medie_rm_low_rad:.2f}, rad >5: {medie_rm_high_rad:.2f}")
print(f"Tasse medie (TAX) - rad <=5: {medie_tax_low_rad:.2f}, rad >5: {medie_tax_high_rad:.2f}")
```

```python
employees_df = pd.DataFrame({ 'employee_id': [101, 102, 103, 104, 105], 
                             'name': ['Alice', 'Bob', 'Charlie', 'David', 'Emma'], 
                             'department_id': [1, 2, 1, 2, 3] })
departments_df = pd.DataFrame({ 'department_id': [1, 2, 3],
                               'department_name': ['HR', 'IT', 'Finance'], 
                               'location': ['New York', 'San Francisco', 'Chicago'] })
import pandas as pd
```

```python
# Unione tramite merge sulla colonna comune 'department_id'
merged_df = employees_df.merge(departments_df, on='department_id')
print(merged_df)
```

```python
# Trasformare department_id in indice
employees_df.set_index('department_id', inplace=True)
departments_df.set_index('department_id', inplace=True)

# Unione tramite join
joined_df = employees_df.join(departments_df)
print(joined_df)

#Quando usi .merge(), puoi specificare la colonna di unione (on='department_id') senza modificare gli indici.
#Quando usi .join(), la funzione utilizza gli indici come chiavi di unione. Se gli indici non sono uguali o non sono stati impostati correttamente, il risultato potrebbe essere diverso o incompleto.
```

```python
import os
import dotenv
import sqlalchemy
import pandas as pd

# Percorso completo del file .env
file_path = r"C:\Users\paster\cred.env"

# Carica le variabili di ambiente dal file specificato
dotenv.load_dotenv(dotenv_path=file_path, override=True)

# Ottieni le variabili di ambiente
username = os.getenv("username")
password = os.getenv("password")
host = os.getenv("host")
dbname = os.getenv("dbname")

# Crea la stringa di connessione
connection_string = f"mysql+pymysql://{username}:{password}@{host}/{dbname}"
print(connection_string)

# Crea l'engine per SQLAlchemy
db_engine = sqlalchemy.create_engine(connection_string)
query="SELECT * FROM dimemployee "
dimemployee=pd.read_sql(query,db_engine)
dimemployee
```

```python
#importo dimemployeesalesterritory
query="SELECT * FROM dimemployeesalesterritory"
dimemployeesalesterritory=pd.read_sql(query,db_engine)
print(dimemployeesalesterritory)
```

```python
# Join su EmployeeKey
df_joined = dimemployee.merge(dimemployeesalesterritory, on='EmployeeKey', how='inner')

# Controlla la dimensione del risultato
print(f"Numero di record dopo il join: {len(df_joined)}")
```

```python
#Carichiamo la tabella dimsalesterritory
query="SELECT * FROM dimsalesterritory"
dimsalesterritory = pd.read_sql(query,db_engine)
dimsalesterritory
```

```python
# Join su SalesTerritoryKey
df_final = df_joined.merge(dimsalesterritory, on='SalesTerritoryKey', how='inner')
df_final
```

```python
print(df_final.head())
```

```python
# Conta per paese
dipendenti_per_paese = df_final['SalesTerritoryCountry'].value_counts()
print("Dipendenti per paese:\n", dipendenti_per_paese)

# Conta per regione
dipendenti_per_regione = df_final['SalesTerritoryRegion'].value_counts()
print("Dipendenti per regione:\n", dipendenti_per_regione)
```

```python
media_baserate_per_paese = df_final.groupby('SalesTerritoryCountry')['BaseRate'].mean().round(2)
print("Media del BaseRate per paese:\n", media_baserate_per_paese)
```
