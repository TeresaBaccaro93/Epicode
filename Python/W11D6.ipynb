```python
import os
import dotenv
import sqlalchemy
import pandas as pd

# Percorso completo del file .env
file_path = r"C:\Users\paster\cred.env"

# Carica le variabili di ambiente dal file specificato
dotenv.load_dotenv(dotenv_path=file_path, override=True)

# Ottieni le variabili di ambiente
username = os.getenv("username")
password = os.getenv("password")
host = os.getenv("host")
dbname = os.getenv("dbname")

# Crea la stringa di connessione
connection_string = f"mysql+pymysql://{username}:{password}@{host}/{dbname}"
print(connection_string)

# Crea l'engine per SQLAlchemy
db_engine = sqlalchemy.create_engine(connection_string)
query="SELECT * FROM dimproduct"
dimproduct=pd.read_sql(query,db_engine)
dimproduct
```

```python
# Arrotonda i valori di DealerPrice alle due cifre decimali
dimproduct['DealerPrice'].round(2)

# Arrotonda al valore intero più vicino
dimproduct['DealerPrice_rounded'].round()

# Limita i valori tra 0 e 1000 usando .clip()
dimproduct['DealerPrice_int'].clip(lower=0, upper=1000)
```

```python
#Modifico la colonna DealerPrice 
dimproduct['DealerPrice_rounded'] = dimproduct['DealerPrice'].round(2)
dimproduct['DealerPrice_int'] = dimproduct['DealerPrice_rounded'].round()
dimproduct['DealerPrice_clipped'] = dimproduct['DealerPrice_int'].clip(lower=0, upper=1000)
```

```python
# aggiornare direttamente la colonna originale, puoi fare:
dimproduct['DealerPrice'] = dimproduct['DealerPrice'].round(2).round().clip(0, 1000)
```

```python
#DataFrame sintetico, che contiene i guadagni mensili di diverse annate, con il seguente codice:
years = 5 
guadagni = pd.DataFrame({"Mese": list("GFMAMGLASOND"*years), 
                         "Anno": np.repeat(list(range(years)), 12), 
                        "Valore": np.random.randint(800, 5000, 12*years)})
print(guadagni)
```

```python
# Somma cumulativa di tutti i guadagni
guadagni['Somma_Cumulativa_Totale'] = guadagni['Valore'].cumsum()
guadagni
```

```python
# Raggruppa per anno e calcola la somma cumulativa all'interno di ogni gruppo
guadagni['Somma_Cumulativa_Per_Anno'] = guadagni.groupby('Anno')['Valore'].cumsum()
guadagni
```

```python
# Estraiamo nome utente e dominio dall'email
email_split = dimcustomer['EmailAddress'].str.split('@', expand=True)
dimcustomer['Email_User'] = email_split[0]
dimcustomer['Email_Domain'] = email_split[1]
dimcustomer
```

```python
# Estraiamo le parti del numero di telefono
# Prima, rimuoviamo eventuali spazi e poi split
dimcustomer['Phone_Parts'] = dimcustomer['Phone'].str.replace(' ', '').str.split(r'[\(\)\-]', expand=False)
dimcustomer
```

```python
# Estraiamo gli indirizzi email che contengono "21"
emails_with_21 = dimcustomer[dimcustomer['EmailAddress'].str.contains('21')]['EmailAddress']
print(emails_with_21 )
```

```python
# 1. Estraiamo tutti gli indirizzi e-mail che contengono "20" o "10"
emails_with_20_or_10 = dimcustomer[dimcustomer['EmailAddress'].str.contains('20|10', na=False)]
print("Indirizzi e-mail che contengono '20' o '10':")
print(emails_with_20_or_10['EmailAddress'])
```

```python
# Calcolare la lunghezza di ogni indirizzo e-mail
dimcustomer['Email_Length'] = dimcustomer['EmailAddress'].str.len()
# Ordinare il DataFrame per lunghezza dell'email in ordine decrescente e prendere i primi 5
top_5_longest_alt = dimcustomer.sort_values(by='Email_Length', ascending=False).head(5)

# Ordinare il DataFrame per lunghezza dell'email in ordine crescente e prendere i primi 5
top_5_shortest_alt = dimcustomer.sort_values(by='Email_Length', ascending=True).head(5)

print("I cinque indirizzi e-mail più lunghi:")
print(top_5_longest_alt['EmailAddress'])

print("I cinque indirizzi e-mail più corti:")
print(top_5_shortest_alt['EmailAddress'])
```

```python
import pandas as pd
facebook=pd.read_csv(r'C:\Users\paster\Downloads\datasets (3)\datasets\beginner_datasets\facebook.csv')
facebook
```

```python
#Converti la colonna 'status_published' in formato datetime
facebook['status_published'] = pd.to_datetime(facebook['status_published'])
print(facebook['status_published'])
```

```python
# Ottieni informazioni specifiche sulle date
facebook['year'] = facebook['status_published'].dt.year
facebook['month'] = facebook['status_published'].dt.month
facebook['day'] = facebook['status_published'].dt.day
facebook['dayofweek'] = facebook ['status_published'].dt.dayofweek  # 0=Lunedì, 6=Domenica
facebook['dayofyear'] = facebook['status_published'].dt.dayofyear
```

```python
facebook.head()
```

```python
# Filtra i post del 2012
posts_2012 = facebook[facebook['status_published'].dt.year == 2012]
posts_2012
```

```python
# Filtra i post di maggio 2018
posts_maggio_2018 = facebook[(facebook['status_published'].dt.year == 2018) & (facebook['status_published'].dt.month == 5)]
posts_maggio_2018
```

```python
# Crea una colonna per il giorno della settimana (0=lunedì, 6=domenica)
facebook['dayofweek'] = facebook['status_published'].dt.dayofweek
facebook['dayofweek']

# seleziona i post nel weekend (sabato=5, domenica=6)
weekend_posts = facebook[facebook['dayofweek'].isin([5, 6])].shape[0]

# Seleziona i post nei giorni feriali (da lunedì a venerdì)
weekday_posts = facebook[~facebook['dayofweek'].isin([5, 6])].shape[0]

print(f"Post nel weekend: {weekend_posts}")
print(f"Post nei giorni feriali: {weekday_posts}")
```

```python
#Troviamo il primo e l'ultimo post pubblicati in ogni anno:
# Raggruppa per anno e trova il primo e l'ultimo post
primo_post_per_anno = facebook.groupby(facebook['status_published'].dt.year)['status_published'].min()
ultimo_post_per_anno = facebook.groupby(facebook['status_published'].dt.year)['status_published'].max()

print("Primo post di ogni anno:\n", primo_post_per_anno)
print("Ultimo post di ogni anno:\n", ultimo_post_per_anno)
```

```python
# Conta i diversi tipi di post
tipi_post = facebook['status_type'].value_counts()

print("Tipi di post e rispettivi conteggi:\n", tipi_post)
```

```python
import os
import pandas as pd

# Percorso alla directory dei dataset
directory_path = r"C:\Users\paster\Downloads\datasets (3)\__MACOSX\datasets\beginner_datasets,encoding="

# Lista dei file nella directory
file_list = os.listdir(directory_path)

# Lista per i dataset con dati nulli
datasets_con_nulli = []

for filename in file_list:
    # Costruisci il percorso completo del file
    filepath = os.path.join(directory_path, filename)

    # Carica il dataset (assumendo siano CSV)
    try:
        df = pd.read_csv(filepath)

        # Verifica se ci sono valori nulli nel dataset
        if df.isnull().values.any():
            datasets_con_nulli.append(filename)
    except Exception as e:
        print(f"Impossibile leggere {filename}: {e}")

# Risultati
print("Dataset con dati nulli:")
for dataset in datasets_con_nulli:
    print(dataset)
    
```

```python
import os
import pandas as pd

# Percorso della cartella con i file CSV
directory_path = r'C:\Users\paster\Downloads\datasets (3)\datasets\beginner_datasets'

# Lista per memorizzare i nomi dei file con dati nulli
files_con_nulli = []

# Ciclo for per esaminare tutti i file nella directory
for filename in os.listdir(directory_path):
    # Seleziona solo i file con estensione .csv e non nascosti
    if filename.endswith('.csv') and not filename.startswith('._'):
        filepath = os.path.join(directory_path, filename)
        # Legge il file CSV in un DataFrame
        df = pd.read_csv(filepath)
        # Crea la maschera booleana dei dati nulli
        null_mask = df.isna()
        # Conta i valori nulli totali nel DataFrame
        total_nulls = null_mask.sum().sum()
        # Se ci sono dati nulli, stampa il nome del file e lo memorizza
        if total_nulls > 0:
            print(f"{filename} contiene {total_nulls} valori nulli.")
            files_con_nulli.append(filename)

# Alla fine, puoi stampare la lista completa dei file con nulli
print("File con dati nulli:", files_con_nulli)
```
